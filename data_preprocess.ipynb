### Diagnosis Age & TMB.Nonsynonymous & Mutation.Count

```python
def fill_na_each_cat(data,category, variable,method):
    '''
    fill na in one column with mean of a specific category
    data: dataframe
    category: str, name of categorical column
    variable: str, name of numerical column that need imputation
    method: {mean,mode}
    '''
    if method == 'mean':
        data[variable] = data.groupby(category)[variable].apply(lambda g: g.fillna(g.mean()))
    elif method == "mode":
        data[variable] = data.groupby(category)[variable].apply(lambda g: g.fillna(g.mode()[0]))
        
```

```python

fill_na_each_cat(df,'Cancer Type Detailed','Mutation Count','mean')
fill_na_each_cat(df,'Cancer Type Detailed','TMB Nonsynonymous','mean')
fill_na_each_cat(df,'Cancer Type Detailed','Diagnosis Age','mean')
```
