```python
# prepare the environment
import pandas as pd
import nltk 
import re
import numpy as np
import itertools
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings('ignore')
```

##**ad 8 processing**##

```python
#import the data
ad8_initial = pd.read_csv("8-tcga.tsv", sep='\t')
#ad8_initial.head(5)
```

```python
#adding nonexist columns as nan
NaN = np.nan
ad8_initial["Survival Stage"] = NaN
ad8_initial["Person Cigarette Smoking History Pack Year Value"] = NaN
ad8_initial["T Stage"] = NaN
```

```python
#select needed columns
ad8 = ad8_initial[['Cancer Type Detailed','Mutation Count','Fraction Genome Altered',
                   'Diagnosis Age','Sex','Survival Stage','T Stage','Somatic Status',
                   'Person Cigarette Smoking History Pack Year Value','TMB (nonsynonymous)']]
```

```python
#rename the columns
ad8.rename(
    columns={"Smoker":"Smoking History",
                "TMB (nonsynonymous)":"TMB Nonsynonymous"}
          ,inplace=True)
ad8
```

##**ad 9 processing**##

```python
#import the data
ad9_initial = pd.read_csv("9-tcga.tsv", sep='\t')
#ad9_initial.head(5)
```

```python
#adding nonexist columns as nan
ad9_initial["Survival Stage"] = NaN
ad9_initial["Fraction Genome Altered"] = NaN
ad9_initial["T Stage"] = NaN
```

```python
#select needed columns
ad9 = ad9_initial[['Cancer Type Detailed','Mutation Count','Fraction Genome Altered',
                   'Diagnosis Age','Sex','Survival Stage','T Stage','Somatic Status',
                   'TMB (nonsynonymous)']]
```

```python
#rename the columns
ad9.rename(
    columns={"TMB (nonsynonymous)":"TMB Nonsynonymous"}
          ,inplace=True)
ad9
```

##**ad 10 processing**##

```python
#import the data
ad10_initial = pd.read_csv("10-tsp.tsv", sep='\t')
#ad10_initial.head(5)
```

```python
#adding nonexist columns as nan
ad10_initial["Survival Stage"] = NaN
ad10_initial["Fraction Genome Altered"] = NaN
ad10_initial["T Stage"] = NaN
```

```python
#select needed columns
ad10 = ad10_initial[['Cancer Type Detailed','Mutation Count','Fraction Genome Altered',
                   'Survival Stage','T Stage','Somatic Status',
                   'TMB (nonsynonymous)']]
```

```python
#rename the columns
ad10.rename(
    columns={"TMB (nonsynonymous)":"TMB Nonsynonymous"}
          ,inplace=True)
ad10
```

##**Concate 3 datasets**##

```python
ad_pt4 = pd.concat([ad8,ad9,ad10])
ad_pt4
```

```python
ad_pt4.to_csv('Combined_ad_data.csv')
```
